## 데이터 과학
</div>

Colab을 통해 인기 있는 Python 라이브러리를 최대한 활용하여 데이터를 분석하고 시각화할 수 있습니다. 아래 코드 셀에서는 <strong>Numpy</strong>를 사용하여 임의의 데이터를 생성하고 <strong>매트플롯립</strong>으로 이를 시각화합니다. 셀을 클릭하면 코드 수정을 바로 시작할 수 있습니다.

Google Drive 계정에서 스프레드시트를 비롯한 데이터를 Colab 메모장으로 가져오거나 GitHub 등의 여러 다른 소스에서 데이터를 가져올 수 있습니다. Colab을 데이터 과학에 활용하는 방법과 데이터 가져오기에 관해 자세히 알아보려면 <a href="#working-with-data">데이터 사용하기</a> 아래 링크를 참조하세요.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------------------------------------
# 0. 이전 단계: 파일 읽기
# (df 변수가 이미 메모리에 있다고 가정)
# 만약 'df' 변수가 없다고 오류가 나면,
# 이 코드 맨 위에 df = pd.read_excel('1005_0_20210522.xls')를 추가하세요.
# -----------------------------------------------------------

# 1. 필요한 컬럼만 선택 (Time과 CGM)
# CBG 컬럼은 이제 예측 모델에 사용하지 않으므로 제외합니다.
data = df[['Date', 'CGM (mg / dl)']].copy()

# 2. 보간(Interpolation) 하기 "전" 빈 값 개수 확인
print("--- 1. 보간(Interpolation) 전 ---")
print(f"CGM 빈 값 개수: {data['CGM (mg / dl)'].isnull().sum()}")

# 3. 빈 값(NaN)을 '선형 보간'으로 채우기
# (끊어진 점과 점 사이를 직선으로 이어줍니다)
data['CGM (mg / dl)'] = data['CGM (mg / dl)'].interpolate(method='linear')

# 4. 보간(Interpolation) "후" 빈 값 개수 확인
print("\n--- 2. 보간(Interpolation) 후 ---")
print(f"CGM 빈 값 개수: {data['CGM (mg / dl)'].isnull().sum()}  (이 값이 0이 되어야 함)")

# 5. 빈 값이 채워진 그래프 다시 그리기
print("\n--- 3. 빈 값이 채워진(보간된) 그래프 ---")
plt.figure(figsize=(20, 6))
plt.plot(data['Date'], data['CGM (mg / dl)'], label='CGM (Gaps Filled)', color='blue')
plt.axhline(y=70, color='r', linestyle='--', label='Hypoglycemia (70 mg/dL)')
plt.title('Patient 1005_0 (After Interpolation)')
plt.xlabel('Time')
plt.ylabel('Glucose (mg/dL)')
plt.legend()
plt.grid(True)
plt.show()
```

```
import pandas as pd
import numpy as np

# -----------------------------------------------------------
# 0. 이전 단계에서 만들어진 'data' 변수를 사용합니다.
# (Time, CGM (mg / dl) 컬럼이 있고, 빈 값이 없는 상태)
# -----------------------------------------------------------

# --- 1. 예측 대상(Target) 정의 ---

# '저혈당' 기준값
LOW_GLUCOSE_THRESHOLD = 70

# '미래' 예측 시점 (30분 뒤)
# (데이터가 5분 간격이므로 30분 = 6 스텝)
N_STEPS_AHEAD = 6

# 'is_low' 컬럼: 현재 혈당이 70 이하이면 1, 아니면 0
data['is_low'] = (data['CGM (mg / dl)'] <= LOW_GLUCOSE_THRESHOLD).astype(int)

# 'future_low' 컬럼 (최종 정답 y):
# "앞으로 30분(6 스텝) 이내에 'is_low'가 한 번이라도 1이 되는가?"
# .rolling(...).max() : 6개 스텝 중 최댓값을 찾음 (하나라도 1이면 1)
# .shift(-N_STEPS_AHEAD) : 미래의 값을 현재 행으로 끌어옴
data['future_low'] = data['is_low'].rolling(window=N_STEPS_AHEAD, min_periods=1).max().shift(-N_STEPS_AHEAD)

# --- 2. 특성(Features) 정의 ---

# '과거' 참고 시점 (지난 60분)
# (60분 = 12 스텝)
N_STEPS_LOOKBACK = 12

# 't-1' (5분 전), 't-2' (10분 전)... 't-12' (60분 전) 컬럼을 자동으로 생성
for i in range(1, N_STEPS_LOOKBACK + 1):
    data[f'cgm_t-{i}'] = data['CGM (mg / dl)'].shift(i)

# --- 3. 문제집 완성 ---

# shift 작업으로 인해 생긴 맨 앞/뒤의 빈 값(NaN)이 있는 행들을 제거
final_data = data.dropna()

# --- 4. 최종 결과물 확인 ---
print(f"--- AI가 학습할 '문제집' (총 {len(final_data)}개의 문제) ---")
print("  (cgm_t-1 ~ t-12 = 문제(X), future_low = 정답(y))")

# 맨 앞 5줄만 출력
print(final_data[['cgm_t-1', 'cgm_t-2', 'cgm_t-3', 'future_low']].head())
```

```
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------------------------------------
# 0. 이전 단계에서 만들어진 'final_data' 변수를 사용합니다.
# -----------------------------------------------------------

# --- 1. '문제(X)'와 '정답(y)' 분리 ---

# '정답' 컬럼 이름
TARGET_COL = 'future_low'

# '문제'가 될 컬럼들 (cgm_t-1 부터 cgm_t-12 까지)
# (N_STEPS_LOOKBACK = 12 였습니다)
feature_cols = [col for col in final_data.columns if col.startswith('cgm_t-')]

X = final_data[feature_cols] # 문제 (과거 60분 혈당)
y = final_data[TARGET_COL]   # 정답 (30분 내 저혈당 여부)

# --- 2. '훈련용'과 '시험용' 데이터 분리 (중요!) ---

# 시계열 데이터는 순서가 중요하므로 절대 섞으면(shuffle=True) 안 됩니다.
# 80%는 훈련용(과거), 20%는 시험용(미래)으로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print(f"--- 1. 훈련 시작 ---")
print(f"훈련용 문제: {len(X_train)}개, 시험용 문제: {len(X_test)}개")

# --- 3. AI 모델 정의 및 훈련 ---

# 랜덤 포레스트 분류기 모델 (나무 100개짜리 숲)
model = RandomForestClassifier(n_estimators=100, random_state=42)

# 훈련 시작 (공부 시작)
model.fit(X_train, y_train)

print(f"\n--- 2. 훈련(공부) 완료 ---")

# --- 4. AI 모델 평가 (시험) ---

# 훈련된 모델로 '시험용' 문제(X_test)를 풀어보게 함
y_pred = model.predict(X_test)

print(f"\n--- 3. AI 모델 시험 결과 ---")

# 4-1. 정확도 (Accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"정확도: {accuracy * 100:.2f}%")

# 4-2. 분류 리포트 (더 자세한 성능)
print("\n[상세 리포트]")
print(classification_report(y_test, y_pred, target_names=['저혈당 아님 (0)', '저혈당 (1)']))

# 4-3. 혼동 행렬 (Confusion Matrix) - AI가 뭘 헷갈렸는지
print("\n[혼동 행렬 (Confusion Matrix)]")
cm = confusion_matrix(y_test, y_pred)
print(cm)

# 혼동 행렬 시각화
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['예측 (아님)', '예측 (저혈당)'],
            yticklabels=['실제 (아님)', '실제 (저혈당)'])
plt.ylabel('Actual (실제)')
plt.xlabel('Predicted (예측)')
plt.title('Confusion Matrix')
plt.show()
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------------------------------------
# 0. 이전 단계에서 만들어진 'X_train', 'y_train' 등을 사용합니다.
# -----------------------------------------------------------

print("--- 1. '저혈당(1)'을 놓치지 않도록 가중치를 준 모델로 재훈련 ---")

# --- 1. AI 모델 재정의 (class_weight='balanced' 옵션 추가) ---
# 데이터가 더 적은 클래스(아마도 '저혈당(1)')에 더 높은 가중치를 부여합니다.
model_balanced = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight='balanced'  # <-- 핵심 옵션!
)

# --- 2. AI 모델 재훈련 ---
model_balanced.fit(X_train, y_train)

print(f"\n--- 2. 재훈련 완료 ---")

# --- 3. AI 모델 재평가 (시험) ---
y_pred_balanced = model_balanced.predict(X_test)

print(f"\n--- 3. AI 모델 (가중치 적용) 시험 결과 ---")

# 3-1. 정확도
accuracy_balanced = accuracy_score(y_test, y_pred_balanced)
print(f"정확도: {accuracy_balanced * 100:.2f}%")

# 3-2. 분류 리포트
print("\n[상세 리포트]")
print(classification_report(y_test, y_pred_balanced, target_names=['저혈당 아님 (0)', '저혈당 (1)']))

# 3-3. 혼동 행렬 (Confusion Matrix)
print("\n[혼동 행렬 (Confusion Matrix)]")
cm_balanced = confusion_matrix(y_test, y_pred_balanced)
print(cm_balanced)

# 3-4. 혼동 행렬 시각화
plt.figure(figsize=(6, 4))
sns.heatmap(cm_balanced, annot=True, fmt='d', cmap='Blues',
            xticklabels=['예측 (아님)', '예측 (저혈당)'],
            yticklabels=['실제 (아님)', '실제 (저혈당)'])
plt.ylabel('Actual (실제)')
plt.xlabel('Predicted (예측)')
plt.title('Confusion Matrix (Balanced)')
plt.show()
```
